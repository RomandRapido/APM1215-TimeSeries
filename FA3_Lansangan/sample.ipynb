```python
import math
import numpy as np
from matplotlib.pylab import *;
def simplin(u0,N):
    un=np.zeros(N+1)
    tt=np.arange(N+1)
    un[0]=u0
    for i in range(N):
        un[i+1]=un[i]*0.97
    return tt,un#,plot(tt,un,':bo')
tt,un=simplin(u0=320,N=1000)
plot(tt,un,':bo')
show()
print("The calculated array is:")
display(un)
```
